# 배열 생성

## 배열 생성 및 초기화

- `zeros()`: 모든 요소를 0으로 초기화

```python
np.zeros(10)
```

- `ones()`: 모든 요소를 1로 초기화

```python
np.ones(10)
```

```python
np.ones((3,3))
```

- `full()`: 모든 요소를 지정한 값으로 초기화

```python
np.full((3,3), 1.23)
```

- `eye()`: 단위행렬 생성
    - 단위행렬: 주대각선의 원소가 모두 1이고 나머지 원소는 0인 정사각 행렬

```python
np.eye(3)
```

- `tri()`: 삼각행렬 생성

```python
np.tri(3)
```

- `empty()`: 초기화되지 않은 배열 생성
    - 초기화되지 않아서 기존 메모리에 존재하는 값이 있음

```python
np.empty(10)
```

- `_like()`: 지정된 배열과 shape가 같은 행렬 생성
    - np.zeros_like()
    - np.ones_like()
    - np.full_like()
    - np.empty_like() 

```python
print(a1)
np.zeros_like(a1)
```

```python
print(a2)
np.ones_like(a2)
```

```python
print(a3)
np.full_like(a3, 10)
```

## 생성한 값으로 배열 생성

- `arange()`: 정수 범위로 배열 생성

```python
np.arange(0,30,2)
```

- `linspace()`: 범위 내에서 균등 간격의 배열 생성

```python
np.linspace(0,1,5)
```

- `logspace()`: 범위 내에서 균등 간격의 로그 스케일 배열 생성

```python
np.logspace(0.1,1,10) # log_10(a1)=0.1,log_10(a10)=1인 사이에 등간격인 10개의 행텍터 log_10(a2)=0.2,log_10(a3)=0.3... 생성
```

## 랜덤값으로 배열 생성

- `random.random()`: 랜덤한 수의 배열 생성 

```python
np.random.random((3,3))
```

- `random.rand()`: 0 이상 1 미만의 랜덤한 수의 배열 생성

```python
np.random.rand(3,3)
```

- `random.randint()`: 일정 구간의 랜덤 정수의 배열 생성

```python
np.random.randint(0,10,(3,3))
```

- `random.normal()`: 정규분포(normal distribution)를 고려한 랜덤한 수의 배열 생성

```python
np.random.normal(0,1,(3,3)) # 평균=0,표준편차=1
```

- `random.randn()`: 표준정규분포(standart normal distribution)를 고려한 랜덤한 수의 배열 생성

```python
np.random.randn(3,3)
```

# 배열 조회

## 배열 속성 정보

```python
def array_info(array):
    print(array)
    print("ndim:",array.ndim)
    print("shape:",array.shape)
    print("dtype:",array.dtype)
    print("size:",array.size)
    print("itemsize(byte):",array.itemsize)
    print("nbytes(size*itemsize):",array.nbytes)
    print("strides:",array.strides)
```

```python
array_info(a1)
```

```python
array_info(a2)
```

```python
array_info(a3)
```

## 인덱싱(Indexing)

```python
print(a1)
print(a1[0])
print(a1[-1])
print(a1[-2])
```

```python
print(a2)
print(a2[0,0])
print(a2[1,1])
print(a2[2,-1])
```

```python
print(a3)
print(a3[0,0,0])
print(a3[1,1,1])
print(a3[2,-1,-1])
```

## 슬라이싱(Slicing)

- `a[start:stop:step]`

```python
print(a1)
print(a1[0:2])
print(a1[0:])
print(a1[:1])
print(a1[::2])
print(a1[::-1])
```

```python
print(a2)
print(a2[1])
print(a2[1,:])
print(a2[:2,:2])
print(a2[:,::-1])
print(a2[1:,::-1])
print(a2[::-1,:])
print(a2[::-1,::-1])
```

## 불리언 인덱싱(Boolean Indexing)

- 배열 각 요소의 선택 여부를 불리언(True/False)로 지정

```python
print(a1)
bi = [False, True, True]
print(a1[bi])
```

```python
print(a2)
bi = np.random.randint(0,2,(3,3),dtype=bool)
print(bi)
print(a2[bi])
```

## 팬시 인덱싱(Fancy Indexing)

- 배열 각 요소의 선택 여부를 인덱스로 지정

```python
print(a1)
idx = [0,2]
print(a1[idx])
idx = np.array([ [0,1],
                 [2,0] ])
print(a1[idx])
```

```python
print(a2)
row = np.array([0,2])
col = np.array([1,2])
print(a2[row,col]) # (0,1),(2,2)
print(a2[row,:])
print(a2[:,col])
print(a2[row,1])
print(a2[row,1:])
print(a2[2,col])
print(a2[1:,col])
```

# 배열 값 삽입/수정/삭제/복사

## 배열 값 삽입

- insert():배열의 특정 위치에 값 삽입
- 원본 배열 변경없이 새로운 배열 반환
- axis를 지정하지 않으면 1차원 배열로 변환 후 삽입

```python
print(a1)
b1 = np.insert(a1,0,10)
print(b1)
```

```python
print(a2)
b2 = np.insert(a2,1,10)
print(b2)
c2 = np.insert(a2,1,10,axis=0)
print(c2)
d2 = np.insert(a2,1,10,axis=1)
print(d2)
```

## 배열 값 수정

```python
print(a1)
a1[0]=1
a1[1]=2
a1[2]=3
print(a1)
a1[:2]=9 # 슬라이싱
print(a1)
idx = np.array([0,2]) # 팬시 인덱싱
a1[idx]=0
print(a1)
a1[idx]+=4
print(a1)
```

```python
print(a2)
a2[0,0]=1
a2[1,1]=2
a2[2,2]=3
print(a2)
a2[0]=1
print(a2)
a2[1:,2]=9 # 슬라이싱
print(a2)
row = np.array([0,1]) # 팬시 인덱싱
col = np.array([1,2])
a2[row,col]=0
print(a2)
```

## 배열 값 삭제

- delete():배열의 특정 위치에 값 삭제
- 원본 배열 변경없이 새로운 배열 반환
- axis를 지정하지 않으면 1차원 배열로 변환 후 삭제

```python
print(a1)
b1 = np.delete(a1,1)
print(b1)
```

```python
print(a2)
b2 = np.delete(a2,1)
print(b2)
c2 = np.delete(a2,1,axis=0)
print(c2)
d2 = np.delete(a2,1,axis=1)
print(d2)
```

## 배열 복사

- 리스트와 달리 슬라이싱해도 `copy()`가 아닌 이상 원본 배열에 영향

```python
print(a2)
a2_sub=a2[:2,:2]
print(a2_sub)
a2_sub[:,1]=0
print(a2_sub)
print(a2)
```

```python
print(a2)
a2_sub_copy=a2[:2,:2].copy()
print(a2_sub_copy)
a2_sub_copy[:,1]=1
print(a2_sub_copy)
print(a2)
```
